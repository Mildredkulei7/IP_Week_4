### 1.45) Uniformity: 

```python
# Procedure :Standardization
# Data Cleaning Action: Fixing messy column names
# Explanation: We change column names hich are uppercase, with spaces, 
# and whitespace all around.
#
autolib_df.columns = autolib_df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
autolib_df
```

```python
# Procedure :Standardization of date and time
# Data Cleaning Action: merging columns with date and timeinto one column.
# Explanation: We merge the year,month, date, hour,minute columns to get one column with datetime to increase accuracy in the results.
#
import datetime

def create_time(row):
  year, month, day, hour, minute = row['year'], row['month'], row['day'], row['hour'], row['minute']
  return datetime.datetime(year,month, day,hour, minute)
  return pd.to_datetime({'year': year, 'month': month,'day': day, 'hour': hour, 'minute': minute})

autolib_df['datetime'] = pd.to_datetime(autolib_df.apply(lambda x: create_time(x), axis=1))

autolib_df.head()
```
